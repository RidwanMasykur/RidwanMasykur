```python
import pandas as pd
import numpy as np
import sklearn

df = pd.read_csv('data_beras.csv')
```

```python
def quality(q_str):
    a = 1
    if q_str == 'Premium':
        a = 3
    elif q_str == 'Medium':
        a = 2

    return a


def month_change(month):
    month_list = ['januari', 'febuari', 'maret', 'april', 'mei', 'juni', 'juli', 'agustus', 'septermber', 'oktober', 'november', 'desember']
    return month_list.index(month) + 1
```

```python
df['bulan'] = df['bulan'].apply(month_change)

df['kualitas'] = df['kualitas'].apply(quality)

df
```

```python
X, Y = df, df['harga']

X = X.drop('harga', axis=1)

X
```

```python
from matplotlib import pyplot as plt

plt.scatter(X['tahun'], Y)
```

```python
plt.scatter(X['kualitas'], Y)
```

```python
plt.scatter(X['bulan'], Y)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, Y)
```

```python
pred = np.array([[2023,3,6]])
predict = model.predict(pred)
```

```python
model.coef_
```

```python
model.rank_
```

```python
model.intercept_
```

```python
model.score(X, Y)
```
